```python
import re
import glob
import sys
import os 

pattern = re.compile(r"\d+\t(.+?)\t_\t\w+\t_\t_\t_\t_\t_\t_")

```

```python
content = ""
with open("../conllu_clean/test/lyrics_cmuc.conllu","r") as file:
    line = file.readline()
    while line:
        #print(line)
        m = re.search(pattern,line)
        if m:
            #print(m.group(1))
            content = content + m.group(1) + "\n"
        if line == "\n":
            #print(line)
            content = content + "\n"
        line = file.readline()
print(content)
```

```python
def get_text_from_conll(input):
    global pattern
    content = ""
    with open(input,"r",encoding="utf-8") as file:
        line = file.readline()
        while line:
            print(line)
            m = re.search(pattern,line)
            if m:
                #print(m.group(1))
                content = content + m.group(1) + " "
            if line == "\n":
                #print(line)
                content = content + "\n"
            line = file.readline()
    return content
```

```python
def get_file_with_text(content,name):
    with open(name,"w") as file:
        file.write(content)
```

```python
contenu = get_text_from_conll("./conllu_clean/test/tweets_cmuc.conllu")
get_file_with_text(contenu,"test.txt")
```
